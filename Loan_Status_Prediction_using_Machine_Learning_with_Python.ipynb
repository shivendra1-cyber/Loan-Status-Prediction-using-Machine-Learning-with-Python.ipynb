importing dependencies

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score

```

data collection processing

```python
#loading dataset to pandas dataframe
loan_dataset = pd.read_csv('/content/train_u6lujuX_CVtuZ9i (1) (1).csv')
```

```python
type(loan_dataset)
```

```python
#printing the first 5 rows of the dataframe
loan_dataset.head()
```

```python
# no of rows and columns
loan_dataset.shape
```

```python
#statistical measures
loan_dataset.describe()
```

```python
#no of missing values in each column
loan_dataset.isnull().sum()
```

```python
# dropping the missing values
loan_dataset = loan_dataset.dropna()
```

```python
#no of missing values in each column
loan_dataset.isnull().sum()
```

```python
# label encoding
loan_dataset.replace({'loan_status':{'N':0,'Y':1}},inplace=True)
```

```python
#printing the first 5 rows of the dataframe
loan_dataset.head()
```

```python
#dependent column values
loan_dataset['Dependents'].value_counts()
```

```python
# replacing the value of 3+ to 4
loan_dataset = loan_dataset.replace(to_replace='3+',value=4)
```

```python
#dependent column values
loan_dataset['Dependents'].value_counts()
```

# data visualisation

```python
# education and loan
sns.countplot(x='Education',hue='Loan_Status',data=loan_dataset)
```

```python
# marital status and loan status
sns.countplot(x='Married',hue='Loan_Status',data=loan_dataset)
```

```python
# convert categorical columns to numerical values
loan_dataset.replace({'Married':{'No':0,'Yes':1},'Gender':{'Female':0,'Male':1},'Self_Employed':{'No':0,'Yes':1},'Property_Area':{'Rural':0,'Semiurban':1,'Urban':2},'Education':{'Not Graduate':0,'Graduate':1},'Loan_Status':{'N':0,'Y':1} },inplace=True)
```

```python
loan_dataset.head()
```

```python
# separating loan and data
X= loan_dataset.drop(columns=['Loan_ID','Loan_Status'],axis=1)
Y=loan_dataset['Loan_Status']
```

```python
print(X)
print(Y)
```

# train test split


```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# training the model : support vectore machine model

```python
classifier=svm.SVC(kernel='linear')
```

```python
# training the support vector machine model
classifier.fit(X_train,Y_train)
```

#model evaluation

```python
# accuracy score on training data
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data:' ,training_data_accuracy)
```

```python
# accuracy score on test data
X_test_prediction=classifier.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on test data:' ,test_data_accuracy)
```
